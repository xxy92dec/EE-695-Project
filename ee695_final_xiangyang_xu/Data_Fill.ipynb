```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("Total_data.csv",index_col = 0)
df.describe()
df = df.drop(['Place', 'Year', 'Race','Sex'],axis = 1)
df.head()
```

# Add new data to NaN 

```python
def Fill_Data(df, s): #df - datafram, s - string of data name
    avg = df[s].mean()
    std = df[s].std()
    nancount = df[s].isnull().sum() #count NaN number
    if avg - std < 0:
        avg = std
    ran = np.random.randint(avg-std, avg+std, size = nancount)
    df[s][np.isnan(df[s])] = ran
```

```python
df.describe()
```

```python
columnValueList = list(df.columns.values)
for i in columnValueList:
    Fill_Data(df, i)
df.describe()   
```

```python
msk = np.random.rand(len(df)) < 0.7
train = df[msk]
test = df[~msk]
```

```python
train.to_csv('Train.csv')
test.to_csv('Test.csv')
```
