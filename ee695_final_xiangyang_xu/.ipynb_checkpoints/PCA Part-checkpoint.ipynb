```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.decomposition import IncrementalPCA
from sklearn import decomposition
import matplotlib.pyplot as plt
from numpy import genfromtxt
```

```python
df = pd.read_csv("New_data.csv",index_col = 0)
y = df.Life_expectancy
X = df.drop('Life_expectancy', axis=1)
```

```python
pca = PCA()
pca.fit(X)
num = 0
count = 1
countlist = []
valuelist = []
for i in range(len(pca.explained_variance_ratio_)):
    num += pca.explained_variance_ratio_[i]
    count += 1
    if num >= 0.99:
        print(count,'principal components account for 99.99% of the covariance matrix')   
        break

num = 0
count = 1
for i in range(len(pca.explained_variance_ratio_)):
    num += pca.explained_variance_ratio_[i]
    valuelist.append(num)
    countlist.append(count)
    count +=1
```

```python
plt.plot(countlist,valuelist)
plt.xlabel('Principal Components')
plt.ylabel('Percentage of the Covariance Matrix Trace')
plt.show()
```

```python
pca = decomposition.PCA(n_components=3)
pc = pca.fit_transform(X)
pclist = pc.tolist()
pc1 = []
pc2 = []
pc3 = []
len(pc)
for i in range(len(pclist)):
    pc1.append(pclist[i][0])
    pc2.append(pclist[i][1])
    pc3.append(pclist[i][2])
```

```python
plt.plot(pc1)
plt.xlabel('PC1')
plt.show()
plt.plot(pc2)
plt.xlabel('PC2')
plt.show()
plt.plot(pc3)
plt.xlabel('PC3')
plt.show()
```
