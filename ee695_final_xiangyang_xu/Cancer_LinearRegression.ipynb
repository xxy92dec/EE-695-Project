```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
def errorCalculate(Origin,Predict):
    errorlist = []
    abserrorlist = []
    error = 0
    count = 0
    errorpercentage = 0

    for i in range(len(Origin)):
        count += 1
        True_num = Origin[i]
        Cal_num = Predict[i]
        Diff = Cal_num - True_num
        absDiff = abs(Cal_num - True_num)
        errorlist.append(Diff)
        abserrorlist.append(absDiff)
    errorlist = np.asarray(errorlist)
    abserrorlist = np.asarray(abserrorlist)
    return errorlist, abserrorlist
```

```python
df = pd.read_csv('train.csv', index_col = 0)
TrainX = df.Cancer
Trainy = df.Life_expectancy
x_array = np.asarray(TrainX)
y_array = np.asarray(Trainy)
```

```python
v = np.polyfit(x_array, y_array, 2)
x = [float(i) for i in range(1, 1000)]
y = [v[0]*i*i + v[1]*i + v[2] for i in range(1,1000)]
plt.plot(x_array,y_array,'r.')
plt.plot(x,y,'b')
plt.show()
```

```python
v = np.polyfit(x_array, y_array, 3)
x = [float(i) for i in range(1, 400)]
y = [v[0]*i*i*i + v[1]*i*i + v[2]*i +v[3] for i in range(1,400)]
plt.plot(x_array,y_array,'r.')
plt.plot(x,y,'b')
plt.show()
```

```python
v = np.polyfit(x_array, y_array, 1)
x = [float(i) for i in range(1, 400)]
y = [v[0]*i + v[1] for i in range(1,400)]
plt.plot(x_array,y_array,'r.')
plt.plot(x,y,'b')
plt.show()
```

```python
df = pd.read_csv('train.csv', index_col = 0)
TestX = df.Cancer
Testy = df.Life_expectancy
x_array = np.asarray(TestX)
y_array = np.asarray(Testy)
Predict = []
for i in x_array:
    predict = v[0]*i + v[1]
    Predict.append(predict)
errorlist, abs_errorlist = errorCalculate(y_array,Predict)
plt.plot(y_array, 'r.')
plt.plot(Predict, 'b.')
plt.title('Red - Original Life Expectancy, Blue - Predict Life Expectancy')
plt.xlabel('Sample ID')
plt.ylabel('Life Expectancy')
plt.show()
plt.plot(errorlist,'g+')
plt.xlabel('Sample ID')
plt.ylabel('Error')
plt.show()
stats.describe(abs_errorlist)
```
