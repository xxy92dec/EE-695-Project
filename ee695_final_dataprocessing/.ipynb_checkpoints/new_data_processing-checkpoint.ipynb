```python
import pandas as pd 
```

# Dataframe making

```python
# df_original
df = pd.read_csv("Life_expentance.csv") 
# df --> List
Row_list =[]   
head_list = ['Place','Year','Sex','Race','Life expectancy']
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    Row_list.append(my_list) 
```

```python
df = pd.read_csv("All_type_of_cancer.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'Cancer')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break                        

df = pd.read_csv("HIV_AIDS_Rate.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'HIV')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break 

df = pd.read_csv("Diabetes.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'Diabetes')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break                        
#Heart_Disease
df = pd.read_csv("Heart_Disease.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'Heart Disease')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break    

#Vehicle
df = pd.read_csv("Vehicle.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'Vehicle')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break    
#Firearm
df = pd.read_csv("Firearm.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'Firearm')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break    
#Pneumonia_and_Influenza_Mortality_Rate
df = pd.read_csv("Pneumonia_and_Influenza_Mortality_Rate.csv") 
list_temp = []
for index, rows in df.iterrows(): 
    my_list =[rows.Place, rows.Year, rows.Sex, rows.Race, rows.Value] 
    list_temp.append(my_list) 
head_list.insert(5,'Pneumonia and Influenza')
for i in range(len(Row_list)):
    Row_list[i].insert(5,'NaN')

for i in range(len(Row_list)):
    for j in range(len(list_temp)):
        if list_temp[j][0] ==  Row_list[i][0]:
            if list_temp[j][1] ==  Row_list[i][1]:
                if list_temp[j][2] ==  Row_list[i][2]:
                    if list_temp[j][3] ==  Row_list[i][3]:
                        Row_list[i][5] = list_temp[j][4]
                        break    
```
